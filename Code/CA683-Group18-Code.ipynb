Data Preprocessing for Cleaned Data

As first step of the preprocessing, we convert all our review data to lower case.

For example - 

Input - Hi! My name is Jane Doe.

Result -  hi! my name is jane doe.

```python
yelp_training['Review_Token'] = yelp_training['Review'].astype(str).str.lower()
```

```python
regexp = RegexpTokenizer('\w+')
```

Next, we tokenize our data to remove special characters and split our words. 

For example -

Input-  hi! my name is jane doe.

Result- ['hi', 'my', 'name', 'is', 'jane', 'doe']

```python
with tf.device('/device:GPU:0'):
  yelp_training['Review_Token'] = yelp_training['Review_Token'].apply(regexp.tokenize)
```

```python
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
```

```python
# Make a list of english stopwords
stopwords = nltk.corpus.stopwords.words("english")
```

Next, we remove stopwords.

For Example -

Input - ['hi', 'my', 'name', 'is', 'jane', 'doe']

Result - ['hi', 'name', 'jane', 'doe']

```python
with tf.device('/device:GPU:0'):
  yelp_training['Review_Token'] = yelp_training['Review_Token'].apply(lambda x: [item for item in x if item not in stopwords])
```

Here we are removing all the words with length less than 2.

Input - ['hi', 'name', 'jane', 'doe']

Result - ['hi', 'name', 'jane', 'doe']

```python
with tf.device('/device:GPU:0'):
  yelp_training['Review_Token'] = yelp_training['Review_Token'].apply(lambda x: ' '.join([item for item in x if len(item)>2]))
```

Sentiment Analysis on Labelled Data

```python
yelp_training = yelp_training.reset_index().rename(columns={'index':'ReviewId'})
```

```python
yelp_training.head()
```

Here we are using TextBlob for calculating sentiment score

```python
def get_sentiment(text):
     scores_dict = {'score' : TextBlob(text).sentiment.polarity }
     return  scores_dict
```

Next we are calculating sentiment score of each review and putting it in a dictionary against ReviewId

```python
res = {}
with tf.device('/device:GPU:0'):
  for i, row in tqdm(yelp_training.iterrows(),total=len(yelp_training)):
    try:
      text = row['Review']
      myid = row['ReviewId']
      score = get_sentiment(str(text))
      res[myid]  = score
    except RuntimeError as e:
      print("Exception Occured while code Execution at: "+ str(i)+ " where: " + str(e))
```

```python
df = pd.DataFrame(res).T
df = df.reset_index().rename(columns={'index':'ReviewId'})
df = df.merge(yelp_training, how = 'left')
```

# TF-IDF

```python
full_result = pd.DataFrame(columns = ['Preprocessing', 'Model', 'Precision', 'Recall', 'F1-score', 'Accuracy'])
```

```python
def tfidf(data, ngrams = 1):

    df_temp = data.copy(deep = True)

    tfidf_vectorizer = TfidfVectorizer(ngram_range=(1, ngrams))
    tfidf_vectorizer.fit(df_temp['Review_Token_lem'])

    list_corpus = df_temp["Review_Token_lem"].tolist()
    list_labels = df_temp["Label"].tolist()

    X = tfidf_vectorizer.transform(list_corpus)
    
    return X, list_labels
```

## ADASYN

```python
def adasyn(X_train, y_train):
  with tf.device('/device:GPU:0'):
    X, y = make_classification(
      n_samples=10000,
      n_features=20,
      n_redundant=0,
      n_informative=10,
      n_classes=2,
      n_clusters_per_class=1,
      weights=[0.9, 0.1]
    )

  nn = NearestNeighbors(n_neighbors=6)
  X_train, y_train = ADASYN(n_neighbors=nn,sampling_strategy='minority').fit_resample(X_train,y_train)

  return X_train, y_train
```

# Naive Bayes - With over sampling - using TF-IDF 

```python
from sklearn import metrics
```

```python
dataN_over = df.copy()
```

```python
with tf.device('/device:GPU:0'):
  X, y = tfidf(dataN_over, ngrams = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
```

```python
X_train, y_train = adasyn(X_train,y_train)
```

```python
with tf.device('/device:GPU:0'):
  clf = MultinomialNB()
  clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)

f1 = f1_score(y_pred = y_pred,y_true= y_test)
pres = precision_score(y_pred = y_pred,y_true= y_test)
rec = recall_score(y_pred = y_pred,y_true= y_test)
acc = accuracy_score(y_pred = y_pred,y_true= y_test)   
```

```python
mse = mean_squared_error(y_test, y_pred)
```

```python
y_pred_proba = clf.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
```

```python
auc = metrics.roc_auc_score(y_test, y_pred_proba)
```

```python
res = {'Preprocessing': 'TF-IDF 1-grams - With ADASYN', 'Model': 'Naive Bayes', 'Precision': pres, 
                     'Recall': rec, 'F1-score': f1, 'Accuracy': acc, 'MSE': mse,'AUC':auc,'FPR':fpr,'TPR':tpr}
result = pd.DataFrame.from_records([res])
full_result =pd.concat([full_result,result], ignore_index = True)
```

```python
#create ROC curve
plt.plot(fpr,tpr,label="AUC="+str(auc))
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend(loc=4)
plt.show()
```
